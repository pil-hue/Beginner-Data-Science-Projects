**Objective:** Exploring a dataset of 200 students, and analysing which factors (namely hours studied, attendance, sleep, previous scores) most strongly influence the final exam scores

```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the file and displaying it
var = pd.read_csv('student_exam_scores.csv')
var
```

```python
var.describe(include = 'all')
```

- The dataset shows a high variation in attendance percentage and previous scores, the rest showing moderate to low variations.
- An average student approximately studies for 6 hours, sleeps for 6.5 hours, and maintains 75% attendance to achieve an average exam score of 34.



```python
var.info()
```

As we can see from above, there are no missing values present in the dataset.

The following scatter plots show the relationship each factor has with the final exam scores.

```python
#Hours Studied Vs Exam Score
plt.scatter(var['hours_studied'], var['exam_score'], color = 'seagreen', alpha = 0.6)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Hours Studied vs Exam Score')
plt.show()
```

Based on the above scatter plot, we find out that Hours Studied has a strong positive correlation with Exam score. This suggests that the more a student studies, the higher their scores tend to be.

```python
#Sleep Hours Vs Exam Score
plt.scatter(var['sleep_hours'], var['exam_score'], color = 'purple', alpha = 0.5)
plt.xlabel('Hours of sleep')
plt.ylabel('Exam Score')
plt.title('Hours of Sleep vs Exam Score')
plt.show()
```

Sleep Hours seem to have a weak correlation with Exam Score as the data points are widely scattered.

```python
#Attendance Percent Vs Exam Score
plt.scatter(var['attendance_percent'], var['exam_score'], color = 'teal', alpha = 0.6)
plt.xlabel('Attendance Percentage')
plt.ylabel('Exam Score')
plt.title('Attendance Percent vs Exam Score')
plt.show()
```

Attendance Percent also shows a weak correlation with exam scores with a lot of points being scattered.

```python
#Previous Exam Scores Vs Final Exam Scores
plt.scatter(var['previous_scores'], var['exam_score'], color = 'red', alpha=0.4)
plt.xlabel("Previous Exam Scores")
plt.ylabel('Exam Scores')
plt.title("Previous Exam Scores vs Final Exam Scores")
plt.show()
```

Previous Exam Scores seems to have a moderate positive correlation since there is an upward trend, however the points are also fairly scattered.

In order to confirm these correlations, we will be using the Pearson's correlation coefficient.

```python
#Finding the Pearson's correlation coefficient for each factor
corr = var.corr(numeric_only=True)['exam_score']
corr = corr.drop(index= 'exam_score')
corr
```

From the above correlation matrix, we can see that the correlations are:

- **Hours Studied vs Exam Score** = 0.77
- **Sleep Hours vs Exam Score** = 0.18
- **Attendance Percent vs Exam Score** = 0.22
- **Previous Scores vs Exam Score** = 0.43

This shows that Hours Studied shows a strong positive correlation with Exam Scores, making it the most influential factor.
Previous Scores shows a moderate positive correlation. On the other hand, Attendance Percent and Sleep Hours show weak positive correlations.

```python
#Bar Graph for each correlation
plt.figure(figsize = (10,5))
plt.bar(corr.index, corr.values)
plt.title('Correlation of each factor with exam score')
plt.xlabel('Factors')
plt.ylabel('Correlation coefficient')

plt.show()
```

The above bar chart further strengthens our conclusion that the hours studied influences the final exam score the most.

**Conclusion:**
The purpose of this analysis was to determine which factor out of all the given four factors influenced a student's exam score the most.

Using the correlation matrix, scatter plots and bar graph, we conclude that:
- The primary influential factor is the number of hours studied, having a correlation coefficient 0.77
- The secondary influential factor is the previous exam scores with coefficient 0.43
- Both sleep hours and attendance percentage showed weak influences on the final score
